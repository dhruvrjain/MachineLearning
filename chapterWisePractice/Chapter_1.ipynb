```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
url="https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
lifesat=pd.read_csv(url)
X=lifesat[["GDP per capita (USD)"]]
Y=lifesat[["Life satisfaction"]]

lm=LinearRegression()
lm.fit(X.values,Y.values)

plt.scatter(X,Y)
plt.plot([25000,59000],[25000*lm.coef_[0][0]+lm.intercept_[0],59000*lm.coef_[0][0]+lm.intercept_[0]],color='r')
plt.xlabel("GDP")
plt.axis([23000,60000,4,9])
plt.ylabel("Life satisfaction")
plt.title("GDP vs Satisfaction")
plt.show()

print(lm.predict([[39000]]))
```
